```python
import sympy as sp
import numpy as np
from IPython.display import display,Latex, Markdown
import math
import matplotlib.pyplot as plt
sp.init_printing("use=latex")
```

```python
# 1.1
v = sp.symbols('v')
a = sp.symbols('a', constant = True)
logistic_sigmoid_fn = 1/(1+sp.exp(-a*v))
logistic_sigmoid_derivative = sp.diff(logistic_sigmoid_fn,v)
display("Derivative value with respect to v:", logistic_sigmoid_derivative)

given_derivative_value = a *logistic_sigmoid_fn *(1-logistic_sigmoid_fn)
display("Derivative value given with respect to v:", given_derivative_value)

assert sp.simplify(logistic_sigmoid_derivative - given_derivative_value) == 0

derivative_value_at_0 = logistic_sigmoid_derivative.subs(v,0)
display("Derivative value at origin:", derivative_value_at_0)

logistic_sigmoid_fn_v_greater_than0 = logistic_sigmoid_fn.subs(v, 1)
display("Logistic sigmoid function value as a tends to infinity (v> 0)", logistic_sigmoid_fn_v_greater_than0.subs(a, sp.oo))


logistic_sigmoid_fn_v_equals_0 = logistic_sigmoid_fn.subs(v, 0)
display("Logistic sigmoid function value as a tends to infinity (v= 0)", logistic_sigmoid_fn_v_equals_0.subs(a, sp.oo))

logistic_sigmoid_fn_exp = sp.simplify(logistic_sigmoid_fn* sp.exp(a*v))
logistic_sigmoid_fn = logistic_sigmoid_fn_exp/sp.exp(a*v)

logistic_sigmoid_fn_v_lesser_than0 = logistic_sigmoid_fn.subs(v, -1)
display("Logistic sigmoid function value as a tends to infinity (v< -1)", logistic_sigmoid_fn_v_lesser_than0.subs(a,sp.oo))

plt.figure(figsize=(16,7))

v_value = np.linspace(-5,5,10000)

a = 1
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'r', label='a=1' )
a = 5
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'b', label='a=5' )
a = 10
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'k', label='a=10' )
a = 100
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'm', label='a=100' )
a = -10
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'c', label='a=-10' )
a = -1
logistic_sigmoid_fn_value = (1-np.exp(-a * v_value)/(1+np.exp(-a* v_value)))
plt.plot(v_value, logistic_sigmoid_fn_value, 'g', label='a=-1' )
plt.xlabel('v, weight')
plt.ylabel('Odd sigmoid function, phi(v)')
plt.legend()
print("As illustrated in the above graph, as the slope parameter,\
'a' value increases the sigmoid function is linear near the origin ")
```

```python
# 1.2
v = sp.symbols('v')
a = sp.symbols('a', constant = True)
# odd_sigmoid_fn = (1-sp.exp(-a*v))/(1+sp.exp(-a*v))
odd_sigmoid_fn = sp.tanh(a*v/2)
odd_sigmoid_derivative = sp.diff(odd_sigmoid_fn,v)
display("Derivative value with respect to v:", (odd_sigmoid_derivative))

given_derivative_value = (a/2) *(1-odd_sigmoid_fn**2)
display("Derivative value given with respect to v:", given_derivative_value)

assert sp.simplify(odd_sigmoid_derivative - given_derivative_value) == 0

derivative_value_at_0 = odd_sigmoid_derivative.subs(v,0)
display("Derivative value at origin:", derivative_value_at_0)

odd_sigmoid_fn_v_greater_than0 = odd_sigmoid_fn.subs(v, 1)
display("Odd sigmoid function value as a tends to infinity (v> 0)", odd_sigmoid_fn_v_greater_than0.subs(a, sp.oo))


odd_sigmoid_fn_v_equals_0 = odd_sigmoid_fn.subs(v, 0)
display("Odd sigmoid function value as a tends to infinity (v= 0)", odd_sigmoid_fn_v_equals_0.subs(a, sp.oo))

odd_sigmoid_fn_exp = sp.simplify(odd_sigmoid_fn* sp.exp(a*v))
odd_sigmoid_fn = odd_sigmoid_fn_exp/sp.exp(a*v)

odd_sigmoid_fn_v_lesser_than0 = odd_sigmoid_fn.subs(v, -1)
display("Odd sigmoid function value as a tends to infinity (v< -1)", odd_sigmoid_fn_v_lesser_than0.subs(a,sp.oo))

plt.figure(figsize=(16,7))

v_value = np.linspace(-5,5,10000)

a = 1
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'r', label='a=1' )
a = 5
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'b', label='a=5' )
a = 10
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'k', label='a=10' )
a = 100
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'm', label='a=100' )
a = -10
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'c', label='a=-10' )
a = -1
odd_sigmoid_fn_value = np.tanh(a*v_value/2.)
plt.plot(v_value, odd_sigmoid_fn_value, 'g', label='a=-1' )
plt.xlabel('v, weight')
plt.ylabel('Odd sigmoid function, phi(v)')
plt.legend()
print("As illustrated in the above graph, as the slope parameter,\
'a' value increases the sigmoid function is linear near the origin ")
```

```python
#1.13-a
x_1,x_2,a,v = sp.symbols('x_1,x_2,a,v')
logistic_sigmoid = (1/(1+sp.exp(-a*v)))
neuron1_output = logistic_sigmoid.subs({v:5*x_1+x_2})
neuron2_output = logistic_sigmoid.subs({v:2*x_1-3*x_2})
neuron3_output = logistic_sigmoid.subs({v:3*neuron1_output-1*neuron2_output})
neuron4_output = logistic_sigmoid.subs({v:4*neuron1_output+6*neuron2_output})
neuron5_output = logistic_sigmoid.subs({v:-2*neuron3_output+neuron4_output})
display("Output of the signal flow graph given with logistic sigmoid:", neuron5_output)
```

```python
#1.13-b
# From 1.1
# Logistic sigmoid slope at origin = a/4
# Linear function = y =mx+c
linear_function_relation = a/4 * (-2*neuron3_output+neuron4_output) + logistic_sigmoid.subs({v:0})
display("Input output relation:", linear_function_relation)
```
