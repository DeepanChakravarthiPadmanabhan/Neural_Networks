## Binary classification Archiv data

## Split data

```python
# Read csv file with the data from Archiv folder
# data=pd.read_csv(r'Data\Archivmixed.csv', delimiter = ',')
data=data_
print(data.shape)#shape of array should be (1478, 1022) I erased the empty columns
data = data.sample(frac=1).reset_index(drop=True)
X = data.iloc[:,0:122].to_numpy()
Y = data.iloc[:,122:].to_numpy()
scaler = MinMaxScaler()
algo = scaler.fit(X)
X= scaler.transform(X)

# Defines ratios, w.r.t. whole dataset.
ratio_train = 0.8
ratio_val = 0.1
test_ratio = 0.1

# Produces test split.
x_remaining, x_test, y_remaining, y_test = skms.train_test_split(
    X, Y, test_size=test_ratio)

# Adjusts val ratio, w.r.t. remaining dataset.
ratio_remaining = 1 - test_ratio
ratio_val_adjusted = ratio_val / ratio_remaining

# Produces train and val splits.
x_train, x_val, y_train, y_val = skms.train_test_split(
    x_remaining, y_remaining, test_size=ratio_val_adjusted)
```

## MLP approach  Define model

```python
class TimeHistory(ke.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.times = []

    def on_epoch_begin(self, batch, logs={}):
        self.epoch_time_start = time.time()

    def on_epoch_end(self, batch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)
        
time_callback = TimeHistory()

model = Sequential()

model.add(Dense(64, activation='relu', input_dim=data.shape[1]-1))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='tanh'))

adam = ke.optimizers.Adam(learning_rate=0.0001,amsgrad=False)
model.compile(loss='binary_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])
model.summary()
desired_epochs=10
history = model.fit(x_train, y_train,
          epochs=desired_epochs,
          batch_size=1,
          validation_data=(x_val, y_val),
          callbacks=[time_callback])
times = time_callback.times


```

```python

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import sympy as sp

# The returned "history" object holds a record
# of the loss values and metric values during training
# print('\nHighest accuracy while training:', np.max(history.history['accuracy']))

# # Predict classes on the test set
y_pred = model.predict_classes(x_test)
y_pred = [-1 if x==0 else 1 for x in y_pred]

cnf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
print(" \n Accuracy of the model after predicting x_test = \n " + str(accuracy))

print("\n Confusion Matrix = \n")
display(sp.Matrix(cnf_matrix))

cnf_report=classification_report(y_test, y_pred)
print("\n Confusion report = \n")
print(cnf_report)

f1_sco=f1_score(y_test, y_pred,average='weighted')
print("\n F1 score = \n")
print(f1_sco)

print("\n Loss per batch")
import matplotlib.pyplot as plt

figure=plt.figure(figsize=(20,10))
plt.subplot(121)
plt.plot(np.linspace(1,desired_epochs,desired_epochs),history.history['loss'])
plt.title("Loss")
plt.xlabel("epoch")
plt.ylabel("loss")

plt.subplot(122)
plt.plot(np.linspace(1,desired_epochs,desired_epochs),history.history['val_loss'])
plt.title("Validation loss")
plt.xlabel("epoch")
plt.ylabel("val_loss")
plt.show()

print("\nTime required to train the neural network \n" +str(np.sum(times))+str('s \n'))

## Compute the memory usage
used_mem = sys.getsizeof(pickle.dumps(history))/1000

print("Used memory(Kb): \n"+ str(used_mem))

local_summary=[]
temp_summary=[cnf_matrix,'mlp',accuracy,f1_sco,np.sum(times),used_mem,"NA"]
local_summary.append(temp_summary)

##Visualize missclassified data 
#predict over all data rows
y_pred = model.predict_classes(X)
y_pred = [-1 if x==0 else 1 for x in y_pred]
misclassified = [i if i == j else 3 for i, j in zip(Y, y_pred)]
counter=0
for each in misclassified:
    if each == 3:
        counter+=1
print(len(misclassified))
print(counter)
```
